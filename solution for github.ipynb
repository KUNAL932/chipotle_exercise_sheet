### Step 1. Import the necessary libraries

```python
import pandas as pd
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv). 

### Step 3. Assign it to a variable called chipo.

```python
chipo = pd.read_csv('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv',sep = '\t')
```

### Step 4. How many products cost more than $10.00?

```python
print(chipo.head())
print(chipo.columns)
```

```python
changer = lambda x : float(x[1:-1])
new_price = chipo.item_price.apply(changer)
new_price.head()
```

```python
chipo.item_price = new_price
chipo_filtered = chipo.drop_duplicates(['quantity','item_name'])
chipo_filtered = chipo_filtered[chipo.quantity == 1]
products_more_than_10 = chipo_filtered[chipo_filtered['item_price'] > 10].item_name.nunique()
products_more_than_10
```

### Step 5. What is the price of each item? 
###### print a data frame with only two columns item_name and item_price

```python
price_column_dataframe = chipo_filtered[['item_name','item_price']]
price_column_dataframe = price_column_dataframe.sort_values(by=['item_price'],ascending = False)
price_column_dataframe.head()
```

### Step 6. Sort by the name of the item

```python
sorted_name = chipo.sort_values(by=['item_name'],ascending = False)
sorted_name
```

### Step 7. What was the quantity of the most expensive item ordered?

```python
most_expensive = chipo.sort_values(by ='item_price',ascending = False)
most_expensive.head(1)
```

### Step 8. How many times were a Veggie Salad Bowl ordered?

```python
veggie_ordered = chipo[chipo.item_name == "Veggie Salad Bowl"]
len(veggie_ordered)
```
